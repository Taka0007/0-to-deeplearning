```python
# import
import numpy as np
import matplotlib.pylab as plt
```

```python
#ステップ関数の定義
def sigmoid(X):
  return 1/(1+np.exp(-X))

X = np.arange(-10.0,10.0,0.1)
Y = sigmoid(X)
plt.plot(X,Y)
plt.ylim(-0.1,1.1)
plt.show()
```
