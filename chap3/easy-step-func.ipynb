```python
# import
import numpy as np
import matplotlib.pylab as plt
```

```python
#ステップ関数の定義
def step_func(X:np.array) -> np.array:
  return np.array( X > 0, dtype=int)

X = np.arange(-10.0,10.0,0.1)
Y = step_func(X)
plt.plot(X,Y)
plt.ylim(-0.1,1.1)
plt.show()
```
