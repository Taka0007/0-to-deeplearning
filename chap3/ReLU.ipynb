```python
# import
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ReLU(X:np.array) -> float:
  return np.maximum(0,X)

X = np.arange(-10.0,10.0,0.1)
Y = ReLU(X)
plt.plot(X,Y)
plt.ylim(-1.0,10.0)
plt.show()
```
